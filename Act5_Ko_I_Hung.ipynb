# Data Manipulation

1. What are the description, finish and price for all products?

```python
# 703
spark.sql('''
SELECT p.productDescription, p.productFinish, p.productPrice
FROM product p
ORDER BY p.productDescription, p.productFinish
''').show()
```

```python
# 704
pd_df.filter(['productDescription', 'productFinish', 'productPrice']
       ).sort_values(['productDescription', 'productFinish'])
```
