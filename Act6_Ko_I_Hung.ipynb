```python
# 704
import pandas as pd
pd_df = pd.read_json('hw1_product.json')
```

#2. What are all details for products with oak finish and price higher than $300 in the descending order of price?

```python
# 703
spark.sql('''SELECT p.*
      FROM Product p
      WHERE p.productFinish = 'Oak' AND p.productPrice > 300
      ORDER BY p.productPrice DESC''').show()
```

```python
# 704
pd_df[(pd_df.productFinish == 'Oak') & (pd_df.productPrice > 300)].sort_values(
    'productPrice', ascending=False)

pd_df[(pd_df.productFinish == 'Oak') & (pd_df.productPrice > 300)]\
  .sort_values('productPrice', ascending=False)
```

```python
pd_df.loc[(pd_df.productFinish == 'Oak') & (pd_df.productPrice > 300)].sort_values(
    'productPrice', ascending=False)
```

```python
pd_df.query('productFinish == \'Oak\' & productPrice > 300').sort_values(
    'productPrice', ascending=False)
```

```python
# 758J
sp_df.filter('productFinish == \'Oak\' and productPrice > 300').sort(
    'productPrice', ascending=False).show()
```

```python
sp_df.filter('productFinish == \'Oak\' and productPrice > 300').sort(
  sp_df.productPrice.desc()).show()
```

```python
sp_df.filter('productFinish == \'Oak\' and productPrice > 300').orderBy(
  sp_df.productPrice.desc()).show()
```

#3. What are the description, finish and price for all products with price higher than $300 in the ascending order of product description then finish?

```python
# 703
spark.sql('''SELECT p.productDescription, p.productFinish, p.productPrice
      FROM Product p
      WHERE p.productPrice > 300
      ORDER BY p.productPrice, p.productFinish ''').show()
```

```python
# 704
pd_df[['productDescription', 'productFinish', 'productPrice']]\
  .loc[(pd_df.productPrice > 300)]\
  .sort_values(['productPrice', 'productFinish'])
```

```python
pd_df.loc[(pd_df.productPrice > 300)]\
   [['productDescription', 'productFinish', 'productPrice']]\
   .sort_values(['productPrice', 'productFinish'])
```

```python
# 758J
sp_df[['productDescription', 'productFinish', 'productPrice']]\
  .filter('productPrice > 300')\
  .orderBy(['productPrice', 'productFinish']).show()
```

```python
sp_df.select('productDescription', 'productFinish', 'productPrice')\
  .filter('productPrice > 300')\
  .orderBy(['productPrice', 'productFinish']).show()
```

```python
sp_df.select('productDescription', 'productFinish', 'productPrice')\
  .filter('productPrice > 300')\
  .orderBy('productPrice', 'productFinish').show()
```

```python
sp_df.filter('productPrice > 300')\
  .select('productDescription', 'productFinish', 'productPrice')\
  .orderBy('productPrice', 'productFinish').show()
```

#4. What are the product finishes?

```python
# 703
spark.sql('''SELECT DISTINCT p.productFinish
       FROM Product p
       ORDER BY p.productFinish''').show()
```
